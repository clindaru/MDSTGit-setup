```python
michigan = df[df['STATE'] == 'MICHIGAN']
michigan_mean = michigan['AVG_READING_4_SCORE'].mean()

ohio = df[df['STATE'] =='OHIO']
ohio_mean = ohio['AVG_READING_4_SCORE'].mean()

print(michigan_mean, ohio_mean)
```

<h2> Data Creation </h2>

_Use this space to create train/test data_

```python
from sklearn.model_selection import train_test_split
```

```python
X = df[['ENROLL','AVG_MATH_8_SCORE','AVG_READING_4_SCORE']].dropna()
y = df.loc[X.index]['AVG_READING_8_SCORE']
```

```python
# X_train, X_test, y_train, y_test = train_test_split(
#      X, y, test_size=, random_state=42)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.3, random_state=0)
```

<h2> Prediction </h2>

ML Models [Resource](https://medium.com/@vijaya.beeravalli/comparison-of-machine-learning-classification-models-for-credit-card-default-data-c3cf805c9a5a)

```python
# import your sklearn class here
from sklearn.linear_model import LinearRegression
```

```python
# create your model here
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

## Evaluation

Choose some metrics to evaluate the performance of your model, some of them are mentioned in the tutorial.

```python
col_name = 'AVG_MATH_8_SCORE'

f = plt.figure(figsize=(12,6))
plt.scatter(X_train['AVG_READING_4_SCORE'], y_train, color = "red")
plt.scatter(X_train['AVG_READING_4_SCORE'], model.predict(X_train), color = "green")

plt.legend(['True Training','Predicted Training'])
plt.xlabel('AVG_READING_4_SCORE')
plt.ylabel('NAME OF THE PREDICTOR')
plt.title("Model Behavior On Training Set")
```
